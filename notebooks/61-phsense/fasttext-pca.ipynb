```python
from sklearn.decomposition import PCA
from matplotlib import pyplot as plt
```

```python
import pickle
with open("gensim_kv_fasttext_tc.pkl", "rb") as fin:
    kv = pickle.load(fin)
```

```python
bisvocab = [(word, vocab.index) for word, vocab in kv.vocab.items() if len(word) == 2]
bisword = np.array([x[0] for x in bisvocab])
bis_vectors = kv.vectors_norm[[x[1] for x in bisvocab], :]
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(bis_vectors)
```

```python
bis_proj = pca.fit_transform(bis_vectors)
```

```python
plt.plot(bis_proj[:, 0], bis_proj[:, 1], '.', markersize=0.2)
```

```python
" ".join(bisword[np.argsort(bis_proj[:, 1])[:20]])
```

```python
" ".join(bisword[np.argsort(bis_proj[:, 1])[-20:]])
```
